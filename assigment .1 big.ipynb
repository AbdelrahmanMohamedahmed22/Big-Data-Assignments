```python
# Start Spark and load your data
from pyspark.sql import SparkSession

spark = SparkSession.builder.getOrCreate()

data = spark.read.option("header", True).option("inferSchema", True) \
    .csv("/content/data.csv")
```

```python
# Employees per city
data.groupBy("City") \
    .count() \
    .orderBy("count", ascending=False) \
    .show()
```

```python
# Average experience per education level
from pyspark.sql.functions import avg

data.groupBy("Education") \
    .agg(avg("ExperienceInCurrentDomain").alias("mean_experience")) \
    .show()
```

```python
# Average age in each city
data.groupBy("City") \
    .agg(avg("Age").alias("average_age")) \
    .show()
```

```python
# Number of female employees with ≥ 2 years experience
female_exp = data.filter(
    (data["Gender"] == "Female") &
    (data["ExperienceInCurrentDomain"] >= 2)
)

print(female_exp.count())
```

```python
# Employees older than 27 who left
left_older = data.filter(
    (data["Age"] > 27) &
    (data["LeaveOrNot"] == 1)
)

print(left_older.count())
```

```python
# Average age for employees with more than 2 years experience
data.filter(data["ExperienceInCurrentDomain"] > 2) \
    .select(avg("Age").alias("avg_age")) \
    .show()
```

```python
# Male employees ( >2 years experience ) who stayed
male_stayed = data.filter(
    (data["Gender"] == "Male") &
    (data["ExperienceInCurrentDomain"] > 2) &
    (data["LeaveOrNot"] == 0)
)

print(male_stayed.count())
```
